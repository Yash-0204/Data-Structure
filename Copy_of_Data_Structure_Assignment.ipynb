Q.2 Explain the difference between mutable and immutable data types with
    examples?

    Mutable data types: The contents or value of the object can be changed (modified) after it is created.
    Example: List, Disctinaries, Sets.

    Immutable data types: Once an object is created, its contents or value cannot be modified. Any operation that would modify it results in a new object being created.
    Example: Strings, Tuples, Integers etc

Q.16 What is a nested dictionary, and give an example of its use case?

     A nested dictionary in Python is simply a dictionary where the values themselves are dictionaries (or other data structures such as lists, tuples, etc.). In other words, it’s a dictionary whose values are other dictionaries, allowing for hierarchical or multi-level data storage.

     Key Characteristics:

     Hierarchical Structure: A nested dictionary can represent complex data with multiple levels of categorization, which is useful for modeling real-world problems or organizing structured data.

     Dynamic and Flexible: Since dictionaries can store any data type as values (including other dictionaries), nested dictionaries offer flexibility in representing complex relationships.

     Accessing Nested Data: To access data within a nested dictionary, you need to use multiple keys, one for each level of nesting.

Q.17 Describe the time complexity of accessing elements in a dictionary?

     The time complexity of accessing elements in a Python dictionary is typically O(1) (constant time), meaning that, on average, dictionary lookups are very fast and do not depend on the size of the dictionary. This is one of the key advantages of dictionaries and makes them ideal for situations where fast lookups are important.

Q.18 In what situations are lists preferred over dictionaries?

     We should prefer lists over dictionaries in these situations:

     1: When the order of elements is important.
     2: When your data is homogeneous and does not need key-value associations.
     3: When you want to sort or modify elements.
     4: When you need to access elements by position (indexing).
